## Which City has the highest customers?

```python
largest_customer_city = supermkt_data['city'].value_counts()
largest_customer_city
```

Yangon had the highest number of customers

```python
largest_customer_city.plot(kind = 'bar', rot = 0, ylabel = 'Number of Customers',
                           xlabel = 'City', title = 'Distribution of Customers', 
                           color = 'blue', figsize = (8,6))
```

## Which gender has the highest count in each of the Cities/Stores?


```python
customer_bygender =  pd.pivot_table(supermkt_data, values= 'total', 
                                    index = 'city', columns = 'gender', 
                                    aggfunc= 'count')
customer_bygender
```

Naypyitaw is dominated by female customers while Yangon is dominated by male customers

```python
from matplotlib.pyplot import legend
customer_bygender.plot(kind = 'bar', rot = 0, ylabel = 'Number of Customers', 
                       xlabel = 'City', title = 'Distribution of Customers by Gender', 
                       figsize = (10,6))
```

## What product has the highest sales?


```python
highest_sales = np.round((pd.pivot_table(supermkt_data, values= 'total', index = 'product_line', aggfunc= np.sum)),2).sort_values(by = 'total', ascending = True)

highest_sales
```

Food and beverage products had the highest total sales amount while health and beauty product generated the least sales amount

```python
highest_sales.plot(kind = 'barh', legend = False, ylabel = 'Total Sales', 
                   xlabel = 'Product Line', title = 'Total Sales by Product Line', 
                   color = 'blue', figsize = (10,6))
```

## What product was bought most male?

```python
# product bought by mostly male customers

most_bought_by_male = supermkt_data[supermkt_data['gender']=='Male'].groupby('product_line').sum('quantity').sort_values(by  = 'quantity')[[ 'quantity']]
most_bought_by_male

```

```python
most_bought_by_male.plot (kind = 'barh', legend = False, xlabel = 'Product Line', 
                          ylabel = 'Quantity', title = 'Distribution of Products Bought by Male Customers',
                          color = 'blue', figsize = (10,6))
```

Male customers bought more health and beauty products thhan any other product category. They bought fashion accessories the least.


## What product was bought most female?

```python
# product bought by mostly female customers

most_bought_by_female = supermkt_data[supermkt_data['gender']=='Female'].groupby('product_line').sum('quantity').sort_values(by  = 'quantity')[[ 'quantity']]
most_bought_by_female
```

```python
most_bought_by_female.plot (kind = 'barh', legend = False, xlabel = 'Product Line', 
                            ylabel = 'Quantity', title = 'Distribution of Products Bought by Female Customers', 
                            color = 'blue', figsize = (10,6))
```

Female customers bought fashion products the most. Health and beauty products were least bought by female customers.


## Which product has the highest average rating?

```python
toprated_product = np.round((supermkt_data.groupby('product_line').mean('rating')[['rating']]),1)
toprated_product.sort_values(by = 'rating')
```

All product categories had approximately close ratings but food and beverages had the most rating of 7.1

```python
toprated_product.sort_values(by = 'rating').plot(kind = 'barh', legend = False, 
                                                 xlabel = 'Product Line', ylabel = 'Rating', 
                                                 title = 'Average Rating of Product Line', 
                                                 color = 'blue', figsize = (10,6))
```

## Which hour of the day has the highest sales?

```python
supermkt_data['date'] = supermkt_data['date'] +' '+ supermkt_data['time']
supermkt_data.drop(['time'], axis=1,inplace=True)
```

```python
supermkt_data['date'] = pd.to_datetime(supermkt_data['date'])  # convert the date to datetime object
supermkt_data['month'] = supermkt_data['date'].dt.month
supermkt_data['day'] = supermkt_data['date'].dt.day
supermkt_data['hour'] = supermkt_data['date'].dt.hour

supermkt_data
```

```python
## highest sales hours
rush_hour = pd.pivot_table(supermkt_data, values= 'invoice_id', index = 'hour',
                           aggfunc= 'count')
rush_hour.sort_values(by = 'invoice_id', ascending = False)
```

The highest sales were made during the 19th hour of the day, that is 7pm.

```python
rushhour_plot = sns.countplot(supermkt_data.hour, color = 'blue')
rushhour_plot.set_xlabel('Hour of the Day')
rushhour_plot.set_ylabel('Number of Sales')
rushhour_plot.set_title('Sales by Hour of Day')
```

```python
rush_hour.plot(kind = 'bar',  rot = 0, legend = False, xlabel = 'Hour of the Day', 
               ylabel = 'Number of Sales', title = 'Sales by Hour of the Day', 
               color = 'blue', figsize = (10,6))
```
